```python
t = AutoTokenizer.from_pretrained('prasanthsagirala/text-to-social-media-captions')
```

```python
m = AutoModelForSeq2SeqLM.from_pretrained('prasanthsagirala/text-to-social-media-captions')
```

```python
import nltk
nltk.download('punkt')
```

```python
text='a woman playing soccer'
inputs = ["captionize: " + text]
inputs = t(inputs, max_length=512, truncation=True, return_tensors="pt")
output = m.generate(**inputs, num_beams=8, do_sample=True, min_length=10, max_length=64)
decoded_output = t.batch_decode(output, skip_special_tokens=True)[0]
predicted_title = nltk.sent_tokenize(decoded_output.strip())[0]
print(decoded_output)
print(predicted_title)
```
